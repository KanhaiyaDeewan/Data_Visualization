## Matplotiib basic

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.linspace( 0,5,11)
x
y = x**3 
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("Chacha")
plt.ylabel("Chachi")
```
